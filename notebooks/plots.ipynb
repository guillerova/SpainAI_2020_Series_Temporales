```python
import matplotlib.pyplot as plt
import seaborn as sns
from data_utils import read_all_candles, read_all_scores
import pandas as pd
```

```python
candles = read_all_candles()
```

```python
candles_close = {k: v[["close"]] for k, v in candles.items()}
candles_close = {k: v.rename(columns = {"close": f"{k}_close"}) for k, v in candles_close.items()}
```

```python
df = pd.concat(candles_close.values(),keys= candles_close.keys(),axis=1)
```

```python
df.columns = df.columns.droplevel()
```

```python
df.plot()
```
